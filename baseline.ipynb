Consumer information

The fields are delimited by "|" and are described as follows:
1. National Drug Code (NDC) in 11 digit normalized format
2. part #  
3. location of image
4. image type - download RxImageImageLayouts.docx to get some visual examples of each type
5. name - description of drug

```python
# read consumer csv file
consumer_df = pd.read_csv("Data/directory_of_images.txt", sep="|", header=None)
# print number of unique values in column 0
print(len(consumer_df[0].unique()))
# change column names to NDC11, part, location, imageType, and name
consumer_df.columns = ["NDC11", "part", "location", "imageType", "name"]
# change data type of column NDC11 to string
consumer_df.NDC11 = consumer_df.NDC11.astype(str)
# fix length of column NDC11 to 11 by adding 0s to the left
consumer_df.NDC11 = consumer_df.NDC11.str.pad(11, side="left", fillchar="0")
# make all drug names upper case
consumer_df.name = consumer_df.name.str.upper()
# split location into 3 columns named "orig_folder", "image", and "file_name"
consumer_df["orig_folder"] = consumer_df.location.str.split("/").str[0]
consumer_df

```
